```python
from pyspark.sql import SparkSession
from pyspark import SparkContext
```

```python
spSession = SparkSession \
            .builder \
            .appName("demo Spark") \
            .getOrCreate()
```

```python
spContext = spSession.sparkContext
```

**cargar datos (csv)**

```python
data = spSession.read.csv('cars.csv', header=True, sep=";")
data.show(5)
```

# **Seleccion de columnas**

```python
from os import truncate
# Forma 1
data.select(data.Car).show(truncate=False) #mejor formato, solo evita que este recortada la descripcion
```

```python
# Forma 2
from pyspark.sql.functions import col
data.select(col('car')).show(truncate=False)
```

```python
# forma 3
data.select(data['Car']).show(truncate=False)
```

```python
data.select(data.Car, data.Cylinders).show(truncate=False)
```

**Agregar una columna**

```python
from pyspark.sql.functions import lit

df = data.withColumn('First_column', lit(1))
df.show(5)
```

```python
df = data.withColumn('Second_column', lit(1))\
      .withColumn('third_column', lit('tercera'))
df.show(5, truncate=False)
```

```python
df.show(5, truncate=False)
```

**Agrupacion**

```python
df.groupBy('Origin').count().show(5, truncate=False)
```

```python
df.groupBy('Horsepower').count().show(5, truncate=False)
```

```python
df.groupBy('Origin', 'Model').count().show(10, truncate=False)
```

```python
df.show(10)
```

**Eliminar**

```python
df = df.drop('Second_column')
df.show()
```

**Ordenar Filas**

```python
df.orderBy('Cylinders').show()
```

```python
df.orderBy('Cylinders', ascending=False).show(truncate=False)
```

```python
df.groupBy('Origin').count().orderBy('count', ascending = False).show(10)
```

```python
total = df.count()
print("Total de registros: ",total)

europa= df.filter(col('Origin') == 'Europe').count()
print("Total de registros Europeos: ",europa)

df.filter(data['origin'] == 'Europe').show(5, truncate = False)
```

```python
europaRegistro= df.filter(col('Origin') == 'Europe').count()
print("Total de registros Europeos: ",europaRegistro)
europaCy= df.filter((col('Origin') == 'Europe')&(data['Cylinders'] == 4)).count()
print("Total de registros Europeos con 4 Cilindros: ",europaCy)
df.filter((data['origin'] == 'Europe')&(data['Cylinders'] == 4)).show(20, truncate = False)
```
